```python
event_dates_all = [event_dates_epf, event_dates_polarT, event_dates_o3, event_dates_anom, event_dates_tend]
```

```python
fig, ax = plt.subplots(figsize=(10, 3))

# Function to plot with vertical offsets so they don't overlap visually
def plot_method(dates, ypos, color, label):
    ax.vlines(dates, ypos - 0.4, ypos + 0.4, color=color, label=label)

plot_method(event_dates_all[0], ypos=3, color='tab:red',   label='epf_reichler')
plot_method(event_dates_all[1], ypos=2, color='tab:blue',  label='polarT_lim (shen et al 2022)')
plot_method(event_dates_all[2], ypos=1, color='tab:green', label='ozone_butler')

plot_method(event_dates_all[3], ypos=4, color='tab:purple', label='u%s_anom_karpechko' % wp0)
plot_method(event_dates_all[4], ypos=5, color='tab:pink', label='u%s_tend_karpechko' % wp0)

# Formatting
ax.set_yticks([1, 2, 3, 4, 5])
ax.set_yticklabels(['ozone_butler', 'polarT_lim (shen et al 2022)', 'epf_reichler', 'u%s_anom_karpechko'% wp0, 'u%s_tend_karpechko'% wp0])
ax.set_xlabel("Year")
ax.set_title("Detected Events - Comparison")
```
